```python
#PROGRAM ANALISIS DASAR SEKUEN DNA
#Menerima sekuens DNA dari pengguna
dna = input("Masukkan sekuens DNA: ").upper().strip()

#Perhitungan Panjang DNA
panjang_dna = len(dna)
#Proses Transkripsi: Mengganti Timin (T) menjadi Urasil (U)
mrna = dna.replace("T", "U")
#Proses Komplemen: Mengganti basa ke pasangan aslinya
#Menggunakan string translation (A-T, C-G)
komplemen = dna.replace('A', 't').replace('T', 'a').replace('C', 'g').replace('G', 'c').upper()

#Menampilkan produk hasil analisis
print("\n=== HASIL ANALISIS DASAR SEKUEN DNA ===")
print("Sekuens DNA Asal    :", dna)
print("Panjang Sekuens     :", panjang_dna, "basa")
print("Produk mRNA         :", mrna)
print("Untai DNA Komplemen :", komplemen)
print("=" * 39)
```

```python
# PROGRAM ANALISIS DASAR SEKUEN DNA (DENGAN KESALAHAN)
# Menerima sekuens DNA dari pengguna
dna = input("Masukkan sekuens DNA: ").uper().strip()

# Perhitungan Panjang DNA
panjang_dna = len(dna)

# Proses Transkripsi: Mengganti Timin (T) menjadi Urasil (U)
mrna = dna.replace("T", "U")

# Proses Komplemen: Mengganti basa ke pasangan aslinya
# Terjadi kesalahan logika karena penggantian langsung (A->T, T->A)
komplemen = dna.replace('A', 'T').replace('T', 'A').replace('C', 'G').replace('G', 'C')

# Menampilkan produk hasil analisis
print("\n=== HASIL ANALISIS DASAR SEKUEN DNA ===")
print("Sekuens DNA Asal     :", dna)
print("Panjang Sekuens      :", panjang_dna "basa")
print("Produk mRNA          :", mrna)
print("Untai DNA Komplemen  :", komplemen)
print("=" * 39)
```
