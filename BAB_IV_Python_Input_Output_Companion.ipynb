## 1) Program structure in Python

Python code is read **top to bottom**.

Core idea:
- *Statements* do work.
- *Blocks* group statements.
- *Indentation* defines blocks (not braces `{}`).


````python
# A tiny program: input → process → output
name = "Ghifaruy"
greeting = "Hello, " + name
print(greeting)

print("Asep", "Bujang", "Cecep", sep=" benci ")
````

## 2) Indentation = structure (critical)

Rule of thumb:
- Indent **4 spaces** per block level.
- Never mix tabs and spaces.

Common block keywords:
- `if`, `elif`, `else`
- `for`, `while`
- `def`, `class`
- `with`, `try`


````python
# Correct indentation
score = 95
if score > 90:
    msg = "Above 90"
else:
    msg = "90 or below"
print(msg)
````

### What breaks when indentation breaks?

Typical error:
- `IndentationError: expected an indented block`

Deliberately uncomment the wrong version to see it.


````python
# Uncomment to test an indentation error
# score = 95
# if score > 90:
# msg = "Above 90"  # <-- missing indentation
# print(msg)
print("Keep the wrong example commented to avoid stopping the notebook.")
````

## 3) Comments (for humans)

- Single-line comment starts with `#`.
- Comments do not affect program output.
- Comments explain *why*, not just *what*.


````python
# This is a comment
x = 10  # Inline comment
print("x =", x)
````

## 4) Escape sequences in strings

Useful ones:
- `\n` newline
- `\t` tab
- `\"` double quote inside a double-quoted string
- `\\` backslash

These control **how text is displayed**.


````python
print("Line 1")
print("Line 2")

print("Line 1\nLine 2")     # explicit newline
print("A\tB\tC")           # tab-separated
print("She said: \"Hello\"")
print("A backslash looks like this: \\")
````

## 5) Output basics: `print()`

`print()` sends text (and other values) to standard output.

Key knobs:
- `sep` : separator between items
- `end` : what to print at the end (default newline)


````python
print("A", "B", "C")
print("A", "B", "C", sep=" RAWR ")
print("No newline at the end...", end="")
print(" <- continues on same line")
````

## 6) Variables & data types (quick practical view)

A variable is a *name* pointing to a value.

Common types:
- `int`, `float`, `str`, `bool`
- containers: `list`, `tuple`, `dict`, `set`

Python is dynamically typed: type is attached to **values**, not variable names.


````python
a = 1        # int
b = 1.5      # float
c = "text"   # str
d = True     # bool

print(type(a), a)
print(type(b), b)
print(type(c), c)
print(type(d), d)
````

## 7) Type conversion (important for input)

`input()` always returns a **string**.

Conversion tools:
- `int("12")` → 12
- `float("3.14")` → 3.14
- `str(99)` → "99"


````python
raw = "12"
print(raw, type(raw))

as_int = int(raw)
as_float = float(raw)
print(as_int, type(as_int))
print(as_float, type(as_float))
````

## 8) Operators (minimum set for I/O tasks)

Arithmetic:
- `+  -  *  /  //  %  **`

Assignment:
- `=  +=  -=  *=  /=  //=  %=  **=`

Comparison:
- `==  !=  >  >=  <  <=`

Logic:
- `and  or  not`


````python
x = 17
y = 5

print("x + y =", x + y)
print("x / y =", x / y)
print("x // y =", x // y)
print("x % y =", x % y)
print("x ** y =", x ** y)

print("x > y:", x > y)
print("x < y: ", x < y)
print("(x > y) and (y == 3):", (x > y) and (y == 3))
````

## 10) String formatting: 3 modern options

### A) f-strings (recommended)

```python
name = "Rani"
score = 92.5
print(f"{name} scored {score:.1f}")
```

### B) `.format()`

### C) `%` formatting (legacy but still seen)


````python
name = "Rani"
score = 92.5

print(f"{name} scored {score:.1f}")           # f-string
print("{} scored {:.1f}".format(name, score)) # format()
print("%s scored %.1f" % (name, score))       # % formatting
````

Typical pattern:

## 13) Defensive input (try/except)

When conversion fails, Python raises `ValueError`.

This pattern keeps the program alive.


````python
def read_int(prompt):
    while True:
        s = input(prompt)
        try:
            return int(s)
        except ValueError:
            print("Not an integer. Try again.")

age = read_int("Enter age (int): ")
print("Age stored:", age)
````

## 17) Exercise 1: Celsius → Kelvin & Fahrenheit

Formulas:
- `K = C + 273.15`
- `F = (9/5)*C + 32`


````python
c = float(input("Celsius (°C): "))
k = c + 273.15
f = (9/5) * c + 32
print(f"{c:.2f} °C = {k:.2f} K")
print(f"{c:.2f} °C = {f:.2f} °F")
````

## 18) Exercise 2: pressure from force and area

Physics reminder:
- `P = F / A`

Try values:
- `F = 5 N`
- `A = 0.0010 m^2`


````python
F = float(input("Force F (N): "))
A = float(input("Area A (m^2): "))

if A != 0:
    P = F / A
    print(f"Pressure P = {P:.2f} Pa")
else:
    print("Area cannot be zero.")
````

## 19) Output quality: clean tables

Often better than many `print()` lines.


````python
items = [
    ("Celcius", 25.0),
    ("Kelvin", 298.15),
    ("Fahrenheit", 77.0),
]

print(f"{'Scale':<12} | {'Value':>10}")
print("-"*26)
for name, val in items:
    print(f"{name:<12} | {val:>10.2f}")
````

## 22) Common pitfalls (fast fixes)

- `ValueError` when converting: input contains letters → use `try/except`.

````python
# Demonstration: string vs numeric addition
a = "2"
b = "3"
print("String addition:", a + b)

a2 = int(a)
b2 = int(b)
print("Numeric addition:", a2 + b2)
````
