## Soal 1 — Validasi Basa DNA (Conditional)

Buat program yang menerima **1 karakter** dari pengguna dan menentukan apakah karakter tersebut termasuk basa DNA yang valid (`A`, `T`, `G`, `C`).

**Tugas:**
1. Gunakan struktur `if-elif-else`.
2. Jika input valid, tampilkan pesan misalnya: `A adalah basa DNA valid`.
3. Jika tidak valid, tampilkan pesan bahwa input bukan basa DNA valid.
4. Program harus menerima huruf kecil maupun huruf besar.

**Contoh input:** `g`  
**Contoh output:** `G adalah basa DNA valid`


**Jawaban:**

```python
# Meminta input 1 karakter dari pengguna
basa = input("Masukkan satu karakter basa DNA: ")

# Validasi panjang input
if len(basa) != 1:
    print("Error: Input >1, Harap masukkan tepat satu karakter.")
else:
    # Mengubah ke huruf besar jika input berupa huruf kecil
    basa_upper = basa.upper()

# Struktur if-elif-else untuk mengecek input yang dilakukan pengguna
if basa_upper == 'A':
    print(f"{basa_upper} adalah basa DNA valid (Adenine)")
elif basa_upper == 'T':
    print(f"{basa_upper} adalah basa DNA valid (Thymine)")
elif basa_upper == 'G':
    print(f"{basa_upper} adalah basa DNA valid (Guanine)")
elif basa_upper == 'C':
    print(f"{basa_upper} adalah basa DNA valid (Cytosine)")
else:
    print(f"{basa} bukan basa DNA valid")
```

**Logika Solusi:**
1. Program menerima input string dari pengguna.
2. Dilakukan validasi awal menggunakan `len()` untuk memastikan input hanya satu karakter.
3. Input dikonversi menjadi huruf besar menggunakan `.upper()` agar perbandingan bersifat *case-insensitive*.
4. Program membandingkan karakter tersebut dengan konstanta basa DNA ('A', 'T', 'G', 'C').

**Alasan Penggunaan Struktur:**
- **`if-elif-else`**: Digunakan karena kita perlu mengevaluasi beberapa kondisi yang saling eksklusif (tidak mungkin satu karakter adalah A sekaligus G). Ini lebih efisien daripada menggunakan beberapa blok `if` terpisah.
- **`.upper()`**: Mempermudah logika perbandingan sehingga kita tidak perlu mengecek huruf kecil secara manual (misal: `basa == 'a' or basa == 'A'`).

## Soal 2 — Klasifikasi Panjang Sekuens (Conditional)

Sebuah laboratorium mengelompokkan panjang sekuens DNA sebagai berikut:
- `< 100` bp → **pendek**
- `100–999` bp → **sedang**
- `>= 1000` bp → **panjang**

Buat program yang menerima input panjang sekuens (integer), lalu menampilkan kategori yang sesuai.

**Tugas tambahan:**
- Jika input negatif, tampilkan pesan error.
- Gunakan struktur percabangan yang benar dan efisien.


**Jawaban:**


```python
# Meminta input panjang sekuens dari pengguna
try:
  panjang = int(input("Masukkan panjang sekuens DNA (bp): "))

# Validasi input negatif
  if panjang < 0:
        print("Error: Panjang sekuens tidak boleh negatif.")
# Klasifikasi menggunakan percabangan efisien
  elif panjang < 100:
        print(f"Kategori: pendek ({panjang} bp)")
  elif panjang <= 999:
        print(f"Kategori: sedang ({panjang} bp)")
  else:
        print(f"Kategori: panjang ({panjang} bp)")
except ValueError:
    print("Error: Harap masukkan angka bulat (integer).")
```

**Logika Solusi:**
1. Program mengambil input dan mengubahnya menjadi tipe data `int`.
2. Pertama, program mengecek apakah nilai kurang dari 0 untuk menangkap input yang tidak valid.
3. Selanjutnya, program mengecek kondisi secara bertingkat: jika tidak < 100, maka otomatis nilainya >= 100, sehingga di `elif` berikutnya kita hanya perlu mengecek batas atasnya saja (<= 999).

**Alasan Penggunaan Struktur:**
- **`if-elif-else`**: Struktur ini adalah struktur yang paling efisien untuk masalah klasifikasi rentang nilai. Begitu satu kondisi terpenuhi, program tidak akan mengecek kondisi di bawahnya, sehingga menghemat proses komputasi.
- **`try-except`**: Digunakan untuk mengantisipasi jika pengguna memasukkan teks atau simbol (bukan angka), sehingga program tidak *crash*.

## Soal 3 — Menghitung Persentase GC (Looping)

Buat program yang menerima sebuah sekuens DNA, lalu menghitung:
1. jumlah `G`
2. jumlah `C`
3. persentase GC

**Rumus:**
$
GC\% = \frac{G + C}{panjang\ sekuens} \times 100
$

**Ketentuan:**
- Gunakan **perulangan** untuk menelusuri karakter pada sekuens.
- Abaikan perbedaan huruf besar dan kecil.
- Jika sekuens kosong, tampilkan pesan yang sesuai.


**Jawaban:**

```python
# Meminta input sekuens DNA dari pengguna
sekuens = input("Masukkan sekuens DNA: ").upper()

# Validasi sekuens kosong
if len(sekuens) == 0:
    print("Error: Sekuens tidak boleh kosong.")
else:
    count_g = 0
    count_c = 0

# Perulangan untuk menelusuri setiap karakter
for basa in sekuens:
    if basa == 'G':
          count_g += 1
    elif basa == 'C':
          count_c += 1

# Menghitung persentase GC
total_gc = count_g + count_c
persentase_gc = (total_gc / len(sekuens)) * 100

# Menampilkan hasil
print(f"Jumlah G: {count_g}")
print(f"Jumlah C: {count_c}")
print(f"Persentase GC: {persentase_gc:.2f}%")
```

**Logika Solusi:**
1. Program meminta input sekuens dan langsung mengubahnya ke huruf besar (`.upper()`).
2. Mengecek apakah sekuens kosong menggunakan `len() == 0` untuk menghindari pembagian dengan nol.
3. Menginisialisasi variabel penghitung `count_g` dan `count_c` mulai dari nol.
4. Melakukan iterasi pada setiap karakter dalam string, menambah hitungan jika ditemukan 'G' atau 'C'.
5. Menghitung persentase berdasarkan perbandingan jumlah (G+C) terhadap total panjang string.

**Alasan Penggunaan Struktur:**
- **`for loop`**: Digunakan untuk mengunjungi setiap karakter dalam string secara efisien satu per satu (iterasi).
- **`if-elif` di dalam loop**: Digunakan untuk memfilter karakter spesifik yang ingin dihitung (G dan C) sambil mengabaikan basa lainnya (A dan T).

## Soal 4 — Menampilkan Kodon per Tiga Basa (Looping)

Buat program yang menerima sekuens DNA, lalu menampilkan kodon-kodon (kelompok 3 basa) secara berurutan.

**Contoh input:** `ATGCCCTAA`  
**Contoh output:**
- Kodon 1: ATG
- Kodon 2: CCC
- Kodon 3: TAA

**Ketentuan:**
- Gunakan **loop**.
- Jika panjang sekuens tidak habis dibagi 3, tampilkan sisa basa di akhir sebagai peringatan.


**Jawaban:**

```python
# Meminta input sekuens dari pengguna
sekuens = input("Masukkan sekuens DNA: ").upper()

# Menghitung jumlah kodon lengkap
jumlah_basa = len(sekuens)
jumlah_kodon = jumlah_basa // 3

# Perulangan untuk mengambil 3 karakter sekaligus
for i in range(0, jumlah_kodon * 3, 3):
    kodon = sekuens[i:i+3]
    nomor_kodon = (i // 3) + 1
    print(f"Kodon {nomor_kodon}: {kodon}")

# Mengecek apakah ada sisa basa yang tidak membentuk kodon utuh
sisa = jumlah_basa % 3
if sisa != 0:
    basa_sisa = sekuens[jumlah_kodon * 3:]
    print(f"\nPeringatan: Sekuens memiliki sisa {sisa} basa di akhir ({basa_sisa}) yang tidak membentuk kodon lengkap.")
```

**Logika Solusi:**
1. Program pertama-tama menangkap input sekuens dan menyeragamkannya menjadi huruf kapital agar tampilannya konsisten.
2. Kita menghitung total panjang string dan menentukan berapa banyak kelompok 3-basa (kodon) utuh yang bisa dibentuk menggunakan operator *floor division* (`//`).
3. Dengan menggunakan fungsi `range(start, stop, step)`, kita melompati indeks string setiap 3 langkah. Di setiap langkahnya, kita melakukan *slicing* string (`sekuens[i:i+3]`) untuk mengambil potongan tiga huruf.
4. Terakhir, program menggunakan operator modulo (`%`) untuk mendeteksi apakah ada sisa karakter di ujung sekuens yang 'tercecer' atau tidak habis dibagi tiga, lalu memberikan peringatan kepada pengguna.

**Alasan Penggunaan Struktur:**
- **`for loop` dengan `range(step=3)`**: Ini adalah pendekatan yang paling efisien. Dibandingkan memeriksa karakter satu per satu, kita bisa langsung melompat ke titik awal setiap kodon. Ini ibarat kita membaca sebuah buku dengan langsung melompati kata per kata, bukan huruf per huruf, jadi prosesnya jauh lebih cepat dan logikanya lebih bersih.
- **String Slicing (`[i:i+3]`)**: Struktur ini dipilih karena kemampuannya yang sangat *pythonic* dalam mengekstrak segmen data. Tanpa perlu membuat loop tambahan di dalamnya, kita bisa langsung 'mengiris' bagian yang kita inginkan dengan presisi tinggi.
- **Modulo (`%`) & Floor Division (`//`)**: Kombinasi dua operator matematika ini memberikan kendali untuk struktur data linear. Ini memastikan bahwa program kita tidak hanya bekerja pada data yang 'sempurna', tetapi juga cerdas dalam mendeteksi anomali atau ketidakteraturan pada panjang data biologis yang seringkali bervariasi.

# Soal 5 — Simulasi Pertumbuhan Bakteri (While Loop)

Jumlah bakteri awal adalah **100 sel**. Setiap iterasi waktu, jumlah bakteri **berlipat dua**.

Buat program yang menghitung berapa kali iterasi diperlukan sampai jumlah bakteri **melebihi 5000 sel**.

**Tugas:**
1. Gunakan `while`.
2. Tampilkan jumlah iterasi.
3. Tampilkan jumlah bakteri pada setiap iterasi.

**Jawaban:**

```python
# Memaskkan nama untuk variabel awal
bakteri = 100
iterasi = 0
target = 5000
print(f"Kondisi Awal: {bakteri} sel\n")

# Perulangan while sampai bakteri melebihi target
while bakteri <= target:
    iterasi += 1
    bakteri *= 2
    print(f"Iterasi {iterasi}: Jumlah bakteri menjadi {bakteri} sel")
print(f"\nHasil Akhir: Diperlukan {iterasi} iterasi untuk melebihi {target} sel.")
```

**Logika Solusi:**
1. Program memulai dengan menetapkan nilai awal populasi sebesar 100 sel dan sebuah penghitung (*counter*) iterasi dari nol.
2. Menggunakan ambang batas 5000 sebagai syarat penghenti.
3. Di dalam setiap putaran, populasi bakteri dikalikan dua (`* 2`) yang meggambarkn proses pembelahan biner, dan penghitung iterasi ditambah satu.
4. Program terus mencetak status populasi di setiap tahap samoai kondisi "lebih dari 5000" terpenuhi, lalu berhenti dan melaporkan total waktu/iterasi yang dibutuhkan.

**Alasan Penggunaan Struktur:**
- **`while loop`**: Struktur ini adalah struktur yang paling tepat digunakan menurut saya, karena kita berada dalam situasi di mana kita tidak tahu pasti berapa banyak langkah yang harus dilakukan sejak awal. Berbeda dengan `for loop` yang biasanya membutuhkan rentang yang pasti, `while` ini bertindak layaknya seorang penjaga pintu yang terus mengizinkan proses berjalan selama syarat 'populasi di bawah target' masih benar. Jadi dengan program ini kita bisa memberikan fleksibilitas dinamis dalam menangani pertumbuhan eksponensial.
- **Sistem Incremental (`iterasi += 1`)**: Kita menggunakan variabel ini bisa disebut sebagai saksi perjalanan waktu. Setiap kali bakteri membelah, variabel ini akan menangkap momen tersebut, jadi di akhir cerita pada programini kita memiliki data yang akurat mengenai durasi pertumbuhan.
- **Operator Penugasan Gabungan (`*= 2`)**: Teknik ini dipilih untuk menjaga kode tetap ringkas tapi tetap memiliki makna yang kuat dalam menggambarkan proses biologis penggandaan diri dengan efisien dalam satu baris instruksi.
